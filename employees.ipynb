**Task: Create 100 Employee Data**
___

This program creates a dataset of `100` unique **employee** records using Python `dictionaries` and `lists`. Each record includes an **employee_id**, **name**, **skills**, **salary**, **designation**, **department**, and **experiences**.

The goal is to practice data generation, nested structures, and save the output as a .json file for use in projects.

```python
# to import random data
import random
# to save .json file
import json

# Sample data
names = ["Uzma", "Sara", "Urwa", "Zaid", "Hassan", "Ali", "Eman", "Zahra", "Bilal", "Riya"]
skills = ["python", "java", "c++", "html", "css", "javascript", "sql", "react", "nodejs", "aws"]
designations = ["engineer", "developer", "manager", "analyst", "consultant", "architect", "admin"]
departments = ["software engineering", "data science", "devops", "QA", "HR", "marketing", "IT support"]

def generate_experience():
    experience_list = []
    for _ in range(2):
        exp_skills = random.sample(skills, random.randint(2, 5))
        experience_list.append({
            "company-id": random.randint(1, 50),
            "designation": random.choice(designations),
            "salary": random.randint(30000, 100000),
            "skills": exp_skills
        })
    return experience_list

def generate_employee(emp_id):
    name = random.choice(names) + str(emp_id)
    base_skills = random.sample(skills, random.randint(2, 4))
    experiences = generate_experience()
    all_skills = set(base_skills)
    for exp in experiences:
        all_skills.update(exp["skills"])
    return {
        "employee_id": emp_id,
        "employee_name": name,
        "employee_skills": list(all_skills),
        "employee_salary": random.randint(50000, 200000),
        "designation": random.choice(designations),
        "department": random.choice(departments),
        "experience": experiences
    }

# Generate 100 employees
employees_data = [generate_employee(i) for i in range(1, 101)]

# Save to JSON file
with open("employee_data_100.json", "w") as f:
    json.dump(employees_data, f, indent=2)

print("File saved as employee_data_100.json")
```
